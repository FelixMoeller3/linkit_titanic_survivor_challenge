```python
train_data = pd.get_dummies(train_data,columns = ['Sex'])
test_data = pd.get_dummies(test_data,columns = ['Sex'])
train_data.head()
```

<h3> 4. Daten vorbereiten </h3>

```python
features = ['Pclass','Age','Sex_female','Sex_male']
target = ['Survived']
#Fare to int
X = train_data[features]
y = train_data[target].astype('category')

X_test = test_data[features]
X_test
```

<h3> 5. Modell trainieren </h3>

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors = 4)#Grid Search 
model.fit(X,y)
prediction = model.predict(X_test)
prediction
```

```python
from sklearn.ensemble import RandomForestClassifier
model2 = RandomForestClassifier(max_depth = 5,n_estimators = 100,random_state=123)
model2.fit(X,y)
prediction2 = model2.predict(X_test)
```

<h3> 6. Abgabe machen </h3>

```python
output = pd.DataFrame({'PassengerId':test_data.PassengerId,'Survived':prediction2})

output.to_csv('Gruppe_x_Heute'+'titanic_prediction.csv',index = False)
```
